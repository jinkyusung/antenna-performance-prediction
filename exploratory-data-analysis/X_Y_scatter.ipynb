```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def x_to_y_visualize(x_column, y_df):
    col = len(y_df.columns)
    fig = plt.figure(figsize=(30, 10 * col))
    
    for i in range(col):
        ax = fig.add_subplot(col, (col+1) // 2, i+1)
        ax.set_title(f'{y_df.columns[i]}', fontdict={'fontsize': 16,'fontweight':'bold'})
        ax.scatter(x_column, y_df.iloc[:, i], s=0.5, c='gray')
    plt.show()
```

```python
import os


print(*list(os.scandir('../dataset/x_feature')), sep='\n', end='\n' * 3)
print(*list(os.scandir('../dataset/y_feature')), sep='\n')
```

```python
x_df = pd.read_csv('../dataset/x_feature/PCB 체결 시 단계별 누름량(Step n).csv')
y_df = pd.read_csv('../dataset/y_feature/신호대 잡음비 (각도n).csv')

print(x_df.columns)
x_to_y_visualize(x_df['X_02'], y_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys


sys.path.append("..")
```

```python
def corr_heatmap(df):
    n = len(df.columns)
    c = df.corr()
    plt.figure(figsize=(n+1, n))
    sns.heatmap(c, annot=True)
```

```python
import os


print(*list(os.scandir('../dataset/x_feature')), sep='\n', end='\n' * 3)
print(*list(os.scandir('../dataset/y_feature')), sep='\n')
```

```python
feature = pd.read_csv('../dataset/x_feature/n번 스크류 삽입 깊이.csv')
corr_heatmap(feature)
```

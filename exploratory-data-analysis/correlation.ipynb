```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys


sys.path.append("..")
```

```python
def corr_heatmap(df):
    n = len(df.columns)
    c = df.corr()
    plt.figure(figsize=(n+1, n))
    sns.heatmap(c, annot=True)
```
